Import ilbraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import dataset

```python
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, 1:].values
print(X)
```

Plot dataset

```python
plt.scatter(X[:,2], X[:,3])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
print(X)
```

Train K-means model

```python
from sklearn.cluster import KMeans
cluster = KMeans()
X_pred = cluster.fit_predict(X)
# print(X_pred)
combined = np.concatenate((X, X_pred.reshape(-1,1)), axis=1)
print(combined)
```

Plot groupings

```python
plt.scatter(X[X_pred==0,2], X[X_pred==0,3], color="red")
plt.scatter(X[X_pred==1,2], X[X_pred==1,3], color="yellow")
plt.scatter(X[X_pred==2,2], X[X_pred==2,3], color="green")
plt.scatter(X[X_pred==3,2], X[X_pred==3,3], color="blue")
plt.scatter(X[X_pred==4,2], X[X_pred==4,3], color="purple")
plt.scatter(X[X_pred==5,2], X[X_pred==5,3], color="cyan")
plt.scatter(X[X_pred==6,2], X[X_pred==6,3], color="magenta")
plt.scatter(X[X_pred==7,2], X[X_pred==7,3], color="brown")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
```

Evaluate model w/ elbow method

```python
# Within-Cluster Sum of Squares
wcss = []
for i in range(1,11):
  cluster = KMeans(n_clusters=i)
  cluster.fit_predict(X)
  wcss.append(cluster.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

Retrain model w/ optimal # of clusters

```python
kmeans = KMeans(n_clusters=5, random_state=0)
X_pred = kmeans.fit_predict(X)
plt.scatter(X[X_pred==0,2], X[X_pred==0,3], color="red")
plt.scatter(X[X_pred==1,2], X[X_pred==1,3], color="pink")
plt.scatter(X[X_pred==2,2], X[X_pred==2,3], color="green")
plt.scatter(X[X_pred==3,2], X[X_pred==3,3], color="blue")
plt.scatter(X[X_pred==4,2], X[X_pred==4,3], color="purple")
print (kmeans.cluster_centers_)
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s = 300, c = 'yellow', label="centroid")
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```
