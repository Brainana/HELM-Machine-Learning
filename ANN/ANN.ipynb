Import libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

Check version of tensorflow installed on the virtual machine

```python
tf.__version__
```

Import the dataset and remove all rows with empty cells

```python
dataset = pd.read_csv("house_prices2.csv")
# Remove rows with NaN values
dataset = dataset.dropna()
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
print(X)
print(y)
```

LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
leColumns = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]
X[leColumns] = X[leColumns].apply(le.fit_transform)
print(X)
```

OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [11])], remainder='passthrough')
# Fit and transform the data
transformed_data = ct.fit_transform(X)
# Convert the transformed data numpy array back to a Pandas DataFrame
X = pd.DataFrame(transformed_data, columns=[
    'furnishingstatus_A','furnishingstatus_B', 'furnishingstatus_C','area', 'bedrooms', 'bathrooms',
                                                         'stories','mainroad', 'guestroom',
                                                         'basement','hotwaterheating',
                                                         'airconditioning', 'parking',
                                                         'prefarea'])
print(X);
```

Split the data into the training and testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print (X_train)
print (y_train)
print(X_test)
print(y_test)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

Initilize the ANN

```python
ann = tf.keras.models.Sequential()
```

Add input layer and first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
# units represents the size of the output
```

Add second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=3, activation="relu"))
```

Add output layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

Compile the ANN

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
# optimizer='adam' indicates gradient descent, and loss='mean_squared_error' means the cost will be evaluated by summing the squares of the differences of the predicted and actual values
```

Train the ANN

```python
print(X_train)
print(y_train)
# batch_size (number of samples per generation) is super important for the training outcome
ann.fit(X_train, y_train, batch_size=64, epochs=3000)
```

Make prediction

```python
prediction = ann.predict(X_test)
# print(prediction)
# print(y_test)
numOfTests = len(X_test)
combine = np.concatenate((prediction, y_test.values.reshape(numOfTests, 1)), 1)
np.set_printoptions(precision=2)
print(combine.astype(int))
```
